# 01 · Explorer les sources agrometflow

Ce notebook construit un tableau simple des variables climatiques et des produits disponibles dans `agrometflow`.


```python
import pandas as pd
from agrometflow.metadata import metadata, products

records = []
for variable, var_info in metadata["climate"].items():
    for product_key, product_info in var_info.get("products", {}).items():
        for source_key, source_info in product_info.get("sources", {}).items():
            records.append(
                {
                    "variable": variable,
                    "product": product_key,
                    "product_name": products.get(product_key, {}).get("name", product_key),
                    "source": source_key,
                    "unit": source_info.get("unit", ""),
                    "resolution": source_info.get("resolution", ""),
                    "description": source_info.get("description", ""),
                }
            )

catalog = pd.DataFrame(records)
catalog.head(12)
```

```python
precip_catalog = catalog[catalog["variable"] == "PR"].sort_values(["product", "source"])
precip_catalog[["variable", "product_name", "source", "unit", "resolution"]]
```

```python
summary = (
    catalog.groupby("product_name")["variable"]
    .nunique()
    .rename("n_variables")
    .sort_values(ascending=False)
    .reset_index()
)
summary
```
