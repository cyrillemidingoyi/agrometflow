## Typologies à retenir

- **Observations in situ** : référence, mais couverture souvent limitée.
- **Télédétection** : couverture large, mais biais dépendants du contexte.
- **Réanalyses** : séries cohérentes et multi-variables, mais résolution parfois trop grossière.
- **Bonne pratique** : croiser les sources plutôt que chercher une source unique parfaite.


```python
import pandas as pd
from agrometflow.metadata import products, list_clim_products, list_sources_for_variable

print("Produits climatiques référencés dans agrometflow:")
print(list_clim_products())
```

```python
product_table = pd.DataFrame(
    [
        {
            "product": key,
            "name": value["name"],
            "description": value["description"],
            "url": value["url"],
        }
        for key, value in products.items()
    ]
).sort_values("product").reset_index(drop=True)

product_table
```

```python
print("Sources disponibles pour la précipitation (PR):")
list_sources_for_variable("PR")
```
