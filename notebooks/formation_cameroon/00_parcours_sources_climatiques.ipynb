## Typologie des donnees

- **Observations in situ** : reference, mais couverture souvent limitee.
- **Teledetection** : couverture large, mais biais dependants du contexte.
- **Reanalyses** : series coherentes et multi-variables, mais resolution parfois trop grossiere.
- **Bonne pratique** : croiser les sources plutot que chercher une source unique parfaite.


## Infrastructures d'acces

Les cellules suivantes montrent les produits et catalogues deja references dans `agrometflow`.
On distingue bien ici l'infrastructure d'acces (catalogues/API) de la typologie des donnees.


```python
import pandas as pd
from agrometflow.metadata import products, list_clim_products, list_sources_for_variable

print("Produits climatiques référencés dans agrometflow:")
print(list_clim_products())
```

```python
product_table = pd.DataFrame(
    [
        {
            "product": key,
            "name": value["name"],
            "description": value["description"],
            "url": value["url"],
        }
        for key, value in products.items()
    ]
).sort_values("product").reset_index(drop=True)

product_table
```
