# 02 · Contrôle qualité journalier

Ce notebook utilise une mini-série construite dans le notebook pour illustrer les flags les plus faciles à expliquer en formation.


## Jeu de données pédagogique

La série contient volontairement:
- une date dupliquée ;
- une pluie journalière très élevée ;
- des répétitions de température ;
- une incohérence simple à commenter.


```python
import pandas as pd
from agrometflow.dataquality.qc import run_qc_pipeline

df = pd.DataFrame(
    [
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 1, "Tx": 31.2, "Tn": 22.1, "rr": 0.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 2, "Tx": 31.2, "Tn": 22.1, "rr": 0.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 3, "Tx": 31.2, "Tn": 22.1, "rr": 0.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 4, "Tx": 31.2, "Tn": 22.1, "rr": 0.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 5, "Tx": 30.8, "Tn": 21.8, "rr": 12.5},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 5, "Tx": 30.8, "Tn": 21.8, "rr": 12.5},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 6, "Tx": 30.5, "Tn": 21.5, "rr": 250.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 7, "Tx": 46.5, "Tn": 22.0, "rr": 5.0},
        {"station": "CM_TRAIN_01", "Year": 2020, "Month": 6, "Day": 8, "Tx": 29.0, "Tn": 21.0, "rr": 3.0},
    ]
)

df
```

```python
qc_results = run_qc_pipeline(
    data=df,
    station_col="station",
    units_map={"Tx": "C", "Tn": "C", "rr": "mm"},
    frequency="daily",
    variable_cols=["Tx", "Tn", "rr"],
)

qc_results["summary"]
```

```python
qc_results["all_flags"].sort_values(["Test", "Var", "Year", "Month", "Day"]).reset_index(drop=True)
```
